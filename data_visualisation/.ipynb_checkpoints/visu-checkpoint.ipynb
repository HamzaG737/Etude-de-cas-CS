```python
import pandas as pd
import pickle
from datetime import datetime

pd.set_option('display.max_colwidth', None)
import seaborn as sns
%matplotlib inline
sns.set_context('poster')
sns.set_color_codes()
```

```python
df = pd.read_csv("df_covid_polarity.csv")
```

```python


def read_probs(probs):      
    """return te probability array with lists"""
    #remove double space and '[' and ']'
    probs = " ".join(probs[1:-1].split())
    return [eval(p) for p in probs.split(" ")]

def read_date(datetime_str):
    return datetime.strptime(datetime_str, '%Y-%m-%d %H:%M:%S')

df["probs"]= [read_probs(p) for p in df["probs"]]
df["datetime"]= [read_date(p) for p in df["datetime"]]
df["date"]= [d.date() for d in df["datetime"]]


```

```python
df["negatif"] = [1 if b else 0 for b in df.predictions==0] #1 if prediction == 0
df["positif"] = [1 if b else 0 for b in df.predictions==1] #1 if prediction == 1
df["neutre"] =  [1 if b else 0 for b in df.predictions==2] #1 if prediction == 2
df["count_dummy"]=1
```

## study emotions 

```python
df_emotions=pickle.load(open('/home/hamza/3A_projets/filiere/saved_data/df_covid_emotions.pkl','rb'))
df_got_emotions=pickle.load(open('/home/hamza/3A_projets/filiere/saved_data/df_got_emotions.pkl','rb'))
```

```python
emotions = ["N", "VALORISATION",
"DEVALORISATION",
"PEUR",
"DESACCORD",
"COLERE",
"ACCORD",
"MEPRIS"]
dict_map = dict(zip(range(len(emotions)),emotions))
df_emotions['emotions'] = df_emotions['predictions'].map(dict_map)
df_got_emotions['emotions'] = df_got_emotions['predictions'].map(dict_map)
```

## stats 

```python
def get_date(datetime_str , got=False):
    if got : 
        return datetime.strptime(datetime_str[:-6], '%Y-%m-%d %H:%M:%S').date()
    return datetime.strptime(datetime_str, '%Y-%m-%d %H:%M:%S').date()
```

```python
df_emotions['date']=df_emotions.datetime.apply(get_date)
df_got_emotions['date']=df_got_emotions.datetime.apply(get_date , got=True)
```

```python
## include got data and transform valo/devalo/n to wanted emotions

wanted_emotions = ["PEUR",
"DESACCORD",
"COLERE",
"ACCORD",
"MEPRIS"]
new_emotions =[]
for index,rows in df_emotions.iterrows():
    probs = rows['probs']
    test=True
    if rows['emotions'] in ['DEVALORISATION','VALORISATION','N'] :
        argmaxes_ = probs.argsort()[-6:][::-1]
        for argmax_ in argmaxes_ : 
            k_best = probs[argmax_]
            if (k_best > 0.05) and (dict_map[argmax_] in  wanted_emotions ) : 
                emotion = dict_map[argmax_] 
                new_emotions.append(emotion)
                test=False
                break
        if test : 
            new_emotions.append(rows['emotions'])
    else : 
        new_emotions.append(rows['emotions'])
new_got_emotions = []
for index,rows in df_got_emotions.iterrows():
    probs = rows['probs']
    test=True
    if rows['emotions'] in ['DEVALORISATION','VALORISATION','N'] :
        argmaxes_ = probs.argsort()[-7:][::-1]
        argmaxes_ = probs.argsort()[-6:][::-1]
        for argmax_ in argmaxes_ : 
            k_best = probs[argmax_]
            if (k_best > 0.05) and (dict_map[argmax_] in  wanted_emotions ) : 
                emotion = dict_map[argmax_] 
                new_got_emotions.append(emotion)
                test=False
                break
        if test : 
            new_got_emotions.append(rows['emotions'])
    else : 
        new_got_emotions.append(rows['emotions'])
            
```

```python
df_emotions_new  = df_emotions.copy()
df_got_emotions_new  = df_got_emotions.copy()

df_emotions_new['emotions'] = new_emotions
df_got_emotions_new['emotions'] = new_got_emotions

all_emotions_new = pd.concat([df_emotions_new[['date','emotions']],df_got_emotions_new[['date','emotions']]],ignore_index=True)
all_emotions_new.emotions.value_counts()
```

```python
#all_emotions = pd.concat([df_emotions[['date','emotions']],df_got_emotions[['date','emotions']]],ignore_index=True)

df_count = pd.merge(all_emotions_new['date'], pd.get_dummies(all_emotions_new['emotions']), right_index = True, 
               left_index = True) 
df_count["count_dummy"]=1
df_agg = df_count.groupby("date").agg("sum")[:-2]
```

```python
from datetime import timedelta 
df = df_agg.reset_index()
k_days = 2
grouped_df = df.groupby(df.index // k_days).mean()

date_end = df['date'][0]
dates = []
total_days =0
while total_days < len(grouped_df) :
    date_begin = date_end 
    date_end = date_begin + timedelta(days=k_days)
    dates.append(str(date_end)[5:] + '--' +str(date_begin)[5:])
    total_days +=1
grouped_df['date'] = dates
grouped_df
```

```python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(24,8))
ax.plot(grouped_df.date, grouped_df.PEUR/grouped_df.count_dummy , color = 'black',marker ='o' , label ='PEUR')
#ax.plot(grouped_df.date, grouped_df.ACCORD/grouped_df.count_dummy , marker ='o' , label ='ACCORD')
ax.plot(grouped_df.date, grouped_df.COLERE/grouped_df.count_dummy , color = 'red',marker ='o' , label ='COLERE')
#ax.plot(df_agg.index, (df_agg.COLERE+df_agg.PEUR)/df_agg.count_dummy , marker ='o' , label ='PEUR + COLERE')
#ax.plot(grouped_df.date, grouped_df.MEPRIS/grouped_df.count_dummy , marker ='o' , label ='MEPRIS')
ax.plot(grouped_df.date, grouped_df.DESACCORD/grouped_df.count_dummy , color='orange',marker ='o' , label ='DESACCORD')

plt.legend(fontsize = 16)
ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))   #to get a tick every 15 minutes
plt.setp(ax.get_xticklabels(), fontsize=14)
plt.setp(ax.get_yticklabels(), fontsize=14)

ax.tick_params(axis='x', rotation=90 )
plt.savefig('emotions.png', transparent=True , bbox_inches='tight')
```
