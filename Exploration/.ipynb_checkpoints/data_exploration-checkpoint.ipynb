```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_context('poster')
sns.set_color_codes()
```

```
from datetime import datetime
def get_time (x) :
  return datetime.strptime(x, '%Y-%m-%d %H:%M:%S').date()

df_covid = pd.read_csv('/content/df_covid_polarity.csv', lineterminator='\n')

df_covid['date'] = df_covid['datetime'].apply(get_time)
```

```
print('first tweet date: ',df_covid['date'].min())
print('last tweet date: ',df_covid['date'].max())
```

```
import matplotlib.dates as mdates

fig,ax = plt.subplots(figsize=(16,8))
df_date = df_covid.groupby('date').count()
ax.plot(df_date.index[:-1], df_date['lang'][:-1] , marker ='o' )
ax.xaxis.set_major_locator(mdates.DayLocator(interval=3))   #to get a tick every 15 minutes
ax.tick_params(axis='x', rotation=90 , size=8)
ax.set_xlabel('Jour')
ax.set_ylabel('Nombre de tweets')
ax.set_title('Evolution du nombre des tweets en fonction du temps')

```

##Wordcloud



```

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

french_stopwords_list =stopwords.words('french') + ['plu','depuis','pays','temps','personnes','jours','jour','beaucoup','pourquoi','dont','nouveau','tous','voir','justes','moins','moment','anne','car','Oui','semaine','toujour','temp','vraiment','personne','face','chez','quoi','sans','toute','juste','aujourd hui','donc','autres','encore','o','soir','fait','bon','faire','trop','cas','non','san','dire','via','comment','faut','rien','dit','entre','peut','dj','avoir','tou','bien','depui','avant','autre','comme','tout','gen','a','quand','moin','ans','aprs','plus','an','ca','alors','si','l','va','tre','trs','peu','cette','aussi']

wordcloud = WordCloud(stopwords=french_stopwords_list,max_words=20, background_color="white").generate(''.join(df_covid['text']))

plt.figure(figsize = (10,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

## Check if there is correlation between parameters and polarity 

```
from scipy import stats

## retweets and polarity 

print('correlation between retweets and polarity : ',stats.pearsonr(df_covid.retweets.values,df_covid.predictions.values )[0])
print('correlation between likes and polarity : ',stats.pearsonr(df_covid.likes.values,df_covid.predictions.values )[0])
print('correlation between replies  and polarity : ',stats.pearsonr(df_covid.replies.values,df_covid.predictions.values )[0])
```
